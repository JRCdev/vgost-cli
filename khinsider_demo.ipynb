```python
userdata = "mario 64"
item = userdata.replace(" ", "+")
print(item)
```

```python
x2 = requests.get("https://downloads.khinsider.com/search?search=" + item)
```

```python
x2.text
```

```python
albumarr = re.findall(r"<a href=\"/game-soundtracks/album.*</a>", x2.text)
print(albumarr)
```

```python
albumlinklib = {}
for x, album in enumerate(albumarr):
    albumlinklib[x] = [album[album.find("\">")+2:-4], "https://downloads.khinsider.com" + album[9:album.find("\">")]]
```

```python
for key in albumlinklib:
    print(key, albumlinklib[key][0])
```

```python
userinput = "15"
albumlink = albumlinklib[int(userinput)][1]
print(albumlink)
```

```python
x3 = requests.get(albumlink)
```

```python
x3.text
```

```python
songarr = re.findall(r"<a href=\"/game-soundtracks/album.*mp3\">[^<]*</a>", x3.text)
print(songarr)
```

```python
songlinklib = {}
for x, song in enumerate(songarr):
    songlinklib[x + 1] = [song[song.find("\">")+2:-4], "https://downloads.khinsider.com" + song[9:song.find("\">")]]
```

```python
for key in songlinklib:
    print(key, songlinklib[key][0], songlinklib[key][1])
```
